## Query 1: Top 5 Customers

```python
SELECT
    customer_id,
    SUM(sales) AS total_sales

FROM orders

GROUP BY customer_id

ORDER BY total_sales DESC

LIMIT 5;
```

Observation:

The query identified the top-performing customers based on total sales generated.

## Query 2: Bottom 5 Customers by Total Sales

```python
SELECT
    customer_id,
    SUM(sales) AS total_sales

FROM orders

GROUP BY customer_id

ORDER BY total_sales ASC

LIMIT 5;
```

Observation:

The query identified customers with the lowest total sales values.

## Query 3: Customers with Only One Order

```python
SELECT
    customer_id,
    COUNT(order_id) AS total_orders

FROM orders

GROUP BY customer_id

HAVING COUNT(order_id) = 1;
```

Observation:

The query displayed customers who placed only one order in the dataset.

## QUERY 4: Customers with Above Average Sales

```python
WITH customer_sales AS (

    SELECT
        customer_id,
        SUM(sales) AS total_sales

    FROM orders

    GROUP BY customer_id

)

SELECT *
FROM customer_sales

WHERE total_sales > (

    SELECT AVG(total_sales)
    FROM customer_sales

)

ORDER BY total_sales DESC;
```

Observation:

The query identified customers whose sales performance was above the average customer sales.

## QUERY 5: Highest Order Value Per Customer

```python

SELECT
    customer_id,
    MAX(sales) AS highest_order_value

FROM orders

GROUP BY customer_id

ORDER BY highest_order_value DESC;
```

Observation:

The query displayed the maximum order value recorded for each customer.
