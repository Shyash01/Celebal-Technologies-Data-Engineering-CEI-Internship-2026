# Data Cleanning

```python
df.isnull().sum()
```

## Missing Value Observation

The dataset contains missing values in several columns.

### Columns with notable missing values:
- `videos` → 781 missing values
- `what_customers_said` → 573 missing values
- `variations` → 562 missing values
- `seller_name` → 301 missing values
- `seller_information` → 301 missing values
- `discount` → 121 missing values

Critical columns such as:
- `title`
- `rating`
- `initial_price`
- `final_price`

do not contain missing values, which is beneficial for further analysis and cleaning operations.

## Missing Value Handling Strategy

Critical columns such as:
- title
- rating
- final_price

do not contain missing values.

Therefore, row deletion is not required.

Instead, selected non-critical columns will be cleaned using appropriate replacement methods.

```python
df['discount'] = df['discount'].fillna(0)
```

```python
df['seller_name'] = df['seller_name'].fillna('Unknown Seller')
```

```python
df['what_customers_said'] = df['what_customers_said'].fillna('No customer review available')
```

```python
df['videos'] = df['videos'].fillna('No video available')
```

```python
df['seller_information'] = df['seller_information'].fillna('Information not available')
```

```python
df['variations'] = df['variations'].fillna('No variations available')
```

```python
df.isnull().sum()
```

## Cleaning Observation

Missing values in important columns were handled using appropriate replacement strategies instead of removing rows.

This approach helps preserve dataset size while improving data consistency for further analysis.

# Price Data Cleaning

```python
df['final_price'].head()
```

```python
df['final_price'] = df['final_price'].str.replace('₹', '', regex=False)
```

```python
df['final_price'] = df['final_price'].str.replace(',', '', regex=False)
```

```python
df['final_price'] = df['final_price'].astype(float)
```

```python
df['final_price'] = df['final_price'].str.replace('"', '', regex=False)
```

```python
df['final_price'] = df['final_price'].astype(float)
```

```python
df['final_price'].dtype
```

## Price Cleaning Observation

The `final_price` column originally contained currency symbols and commas, which caused it to be stored as text data.

The column was cleaned and converted into numerical format to enable calculations and analysis.

```python
high_rated_products = df[df['rating'] > 4]

high_rated_products.head()
```

# Derived Column Creation

```python
df['discount_amount'] = df['initial_price'] - df['final_price']
```

```python
df[['initial_price', 'final_price', 'discount_amount']].head()
```

```python
df['discount_amount'].describe()
```

```python
df[['initial_price', 'discount', 'final_price', 'discount_amount']].sample(10)
```

## Derived Column Observation

A new column named `discount_amount` was created using:

discount_amount = initial_price - final_price

This derived column represents the effective discount value received on each product.

### Key Findings:
- Some products show significant discount values.
- Certain products have zero discount amount, indicating either:
  - no active discount, or
  - pricing inconsistencies in the dataset.
- The derived column helps in understanding product pricing behavior more effectively.
