## 11. Create Spark Session

## 12. Create Existing Dataset - Target Table

```python
target_data = [
    (101, "Rahul", "Delhi"),
    (102, "Aman", "Mumbai")
]

target_df = spark.createDataFrame(
    target_data,
    ["CustomerID", "Name", "City"]
)

target_df.show()
```

## 13. Create New Dataset - Source Table

```python
source_data = [
    (101, "Rahul", "Noida"),
    (103, "Karan", "Pune")
]

source_df = spark.createDataFrame(
    source_data,
    ["CustomerID", "Name", "City"]
)

source_df.show()
```

## 14. SPARK MERGE Logic

Step 1. Register Temp Views

```python
target_df.createOrReplaceTempView("target")
source_df.createOrReplaceTempView("source")
```

```python
spark
```

Step 2. Register Temp Views

```python
target_df.createOrReplaceTempView("target")
source_df.createOrReplaceTempView("source")
```

Step 3. Execute MERGE Logic

```python
merged_df = spark.sql("""
SELECT CustomerID, Name, City
FROM source

UNION ALL

SELECT t.CustomerID, t.Name, t.City
FROM target t
WHERE t.CustomerID NOT IN (
    SELECT CustomerID FROM source
)
""")
```

Step 4. Verify

```python
merged_df.show()
```

## 15. Observation

The merge logic which i used, successfully updated existing records and inserted new records from the source dataset. Spark SQL operations helped combine old and new data efficiently without manually modifying individual records.

This demonstrated the concept of UPSERT (Update + Insert), which is commonly performed using Delta Lake MERGE operations in modern data engineering workflows.
