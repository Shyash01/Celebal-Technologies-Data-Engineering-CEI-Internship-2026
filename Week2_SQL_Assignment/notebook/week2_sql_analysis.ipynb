# SQL-Based Sales Analysis using Superstore Dataset

## Objective
Analyze sales data using SQL with filtering, aggregation, sorting, and business queries using PostgreSQL.

## Tools Used

- PostgreSQL
- pgAdmin 4
- SQL
- Python
- Pandas
- SQLAlchemy
- Jupyter Notebook

# Database Preparation

The dataset was cleaned using Python Pandas and imported into PostgreSQL for analysis. 
Data types were standardized for numerical and date columns before performing analytical queries.

# SQL Filtering Operations

## 1. Sales Records from West Region

```python
SELECT *
FROM superstore
WHERE "Region" = 'West';
```

The West region contains a significant number of sales transactions and customer orders

## 2. Technology category products

```python
SELECT *
FROM superstore
WHERE "Category" = 'Technology';
```

### Observation

Technology products contribute significantly to overall sales performance.

## 3. Orders with Sales Greater Than 500

```python
SELECT *
FROM superstore
WHERE "Sales" > 500;
```

### Observation

High-value orders help identify premium products and bulk purchases.

## 4. Orders Placed in Year 2017

```python
SELECT *
FROM superstore
WHERE "Order Date"
BETWEEN '2017-01-01' AND '2017-12-31';
```

### Observation

Date-based filtering helps analyze yearly sales activity and business growth.

## 5. Technology Products in West Region with Sales Above 500

```python
SELECT *
FROM superstore
WHERE "Category" = 'Technology'
AND "Region" = 'West'
AND "Sales" > 500;
```

### Observation

Applying multiple conditions helps identify high-performing sales segments and profitable markets.

# GROUP BY and Aggregation Operations

## 1. Total Sales by Category

```python
SELECT
    "Category",
    SUM("Sales") AS total_sales
FROM superstore
GROUP BY "Category"
ORDER BY total_sales DESC;
```

### Observation

The analysis shows which product categories contribute the highest revenue to the business.

## 2. Average Profit by Region

```python
SELECT
    "Region",
    AVG("Profit") AS average_profit
FROM superstore
GROUP BY "Region"
ORDER BY average_profit DESC;
```

### Observation

Certain regions generate higher average profit, indicating stronger business performance.

## 3. Total Quantity Sold by Category

```python
SELECT
    "Category",
    SUM("Quantity") AS total_quantity_sold
FROM superstore
GROUP BY "Category"
ORDER BY total_quantity_sold DESC;
```

### Observation

This query identifies categories with the highest product movement and customer demand.

## 4. Number of Orders per Region

```python
SELECT
    "Region",
    COUNT(*) AS total_orders
FROM superstore
GROUP BY "Region"
ORDER BY total_orders DESC;
```

### Observation

The query helps determine regions with the highest customer activity and order volume.

## 5. Average Discount by Category

```python
SELECT
    "Category",
    AVG("Discount") AS average_discount
FROM superstore
GROUP BY "Category"
ORDER BY average_discount DESC;
```

### Observation

Different product categories receive varying discount strategies to improve sales performance.

# Sorting and Limiting Operations

## 1. Top 10 Highest Sales

```python
SELECT
    "Product Name",
    "Sales"
FROM superstore
ORDER BY "Sales" DESC
LIMIT 10;
```

### Observation

The query identifies products generating the highest sales revenue.

## 2. Top Categories by Profit

```python
SELECT
    "Category",
    SUM("Profit") AS total_profit
FROM superstore
GROUP BY "Category"
ORDER BY total_profit DESC;
```

# Business Use Cases

## 1. Monthly Sales Trend

```python
SELECT
    DATE_TRUNC('month', "Order Date") AS month,
    SUM("Sales") AS monthly_sales
FROM superstore
GROUP BY month
ORDER BY month;
```

## 2. Top Customers

```python
SELECT
    "Customer Name",
    SUM("Sales") AS total_sales
FROM superstore
GROUP BY "Customer Name"
ORDER BY total_sales DESC
LIMIT 10;
```

## 3. Duplicate Check

```python
SELECT
    "Order ID",
    COUNT(*)
FROM superstore
GROUP BY "Order ID"
HAVING COUNT(*) > 1;
```

# Data Validation

## 1. Row Count

```python
SELECT COUNT(*)
FROM superstore;
```

## 2. Null Check

```python
SELECT
    COUNT(*) FILTER (WHERE "Sales" IS NULL) AS missing_sales,
    COUNT(*) FILTER (WHERE "Profit" IS NULL) AS missing_profit
FROM superstore;
```
